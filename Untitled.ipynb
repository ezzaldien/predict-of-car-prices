```python
import os
import requests

def download_dataset(url, destination_folder):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Extract the filename from the URL
        filename = os.path.basename(url)
        
        # Create the destination path including the filename
        destination = os.path.join(destination_folder, filename)
        
        # Open the destination file in binary write mode
        with open(destination, 'wb') as f:
            # Write the content of the response to the file
            f.write(response.content)
        print("Dataset downloaded successfully.")
        return destination  # Return the path where the dataset is saved
    else:
        print("Failed to download dataset. Status code:", response.status_code)
        return None

# URL of the dataset
dataset_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"  # Replace this with the actual URL of the dataset

# Destination folder where you want to save the dataset
destination_folder = "D:\paython_vsc\python for data analysis\dataset"  # Change this to your desired destination

# Download the dataset
downloaded_file_path = download_dataset(dataset_url, destination_folder)
if downloaded_file_path:
    print("Dataset saved at:", downloaded_file_path)
else:
    print("Failed to save dataset.")
```
